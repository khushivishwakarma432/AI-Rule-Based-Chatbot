```python
import pandas as pd
import random
from datetime import datetime

# Ask username
username = input("Enter your name: ")

print(f"\n🤖 Hello {username}! I am your AI chatbot.")
print("Type 'bye' to exit.\n")

# Knowledge base
data = {
    "question": [
        "hello",
        "hi",
        "how are you",
        "what is ai",
        "what is machine learning",
        "what is python",
        "who created you",
        "what can you do",
        "bye"
    ],

    "response": [
        ["Hello {username}! 👋", "Hi {username}! Nice to meet you."],

        ["Hi there {username}!", "Hello {username}!"],

        ["I am doing great {username}!", "All good here {username}!"],

        ["AI means Artificial Intelligence.",
         "Artificial Intelligence enables machines to learn and think."],

        ["Machine Learning is a subset of AI.",
         "ML helps computers learn from data."],

        ["Python is a popular programming language used in AI and data science.",
         "Python is widely used for machine learning and automation."],

        ["I was created using Python for a college project.",
         "A student created me using Python."],

        ["I can answer basic AI and programming questions.",
         "I can chat with you and answer simple questions."],

        ["Goodbye {username}! Have a great day."]
    ]
}

# Convert to dataframe using pandas
df = pd.DataFrame(data)

# Chatbot loop
while True:

    user_input = input("You: ").lower()

    match = df[df["question"] == user_input]

    if not match.empty:

        responses = match.iloc[0]["response"]

        reply = random.choice(responses)

        reply = reply.replace("{username}", username)

    else:

        reply = f"Sorry {username}, I don't understand."

    time = datetime.now().strftime("%H:%M:%S")

    print(f"[{time}] Bot: {reply}")

    if user_input == "bye":
        break
```
